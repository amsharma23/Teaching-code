```python
import numpy as np
import matplotlib.pyplot as plt

def generate_sine_sum(max_freq):
    # Number of points in the time domain
    num_points = 1000

    # Time values
    t = np.linspace(0, 2, num_points)
    #print(t)
    # Create subplots
    rows = 3
    cols = 4
    fig, axs = plt.subplots(rows, cols, figsize=(15, 10))
    
    # Flatten the axs array for ease of indexing
    axs = axs.flatten()
    
    for i in range(1, max_freq + 1, 10):
        print(i)
        signal = np.zeros((1, num_points))
        
        # Generate the sum of sine waves
        if(i != 1): 
            for freq in range(1, i, 2):
                signal += (1 / (freq)) * np.sin(np.pi * (freq) * t)
        else:
            signal += np.sin(np.pi * t)
        # Plot the signal in the corresponding subplot
        ax = axs[(i-1)//10]
        ax.plot(t, signal[0], label=f'Max Frequency: {i}')
        ax.set_title(f'Sum of Sine Waves (Max Frequency: {i})')
        ax.set_xlabel('Time')
        ax.set_ylabel('Amplitude')
        ax.legend()
    
    # Adjust layout for better visualization
    plt.tight_layout()
    plt.show()

# Set the desired max_freq value
max_freq_value = 111

# Generate subplots for the specified max_freq value
generate_sine_sum(max_freq_value)
```
